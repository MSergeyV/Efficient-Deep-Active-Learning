```python
from matplotlib import image
from matplotlib.pyplot import imread
from IPython import display

import os
import time
import math
from copy import deepcopy
from tqdm import tqdm_notebook

# Visualization
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
num_pool_points_ar = np.load("num_pool_points_ar.npy").tolist()
dimensions_ar = np.load("dimensions_ar.npy").tolist()

time_simple_results = np.load("time_simple_results.npy").item()
time_fast_results = np.load("time_fast_results.npy").item()
new_approach_time_fast_results = np.load("new_approach_time_fast_results.npy").item()
```

```python
fig = plt.figure(figsize=(20, 10))
    
ax1 = fig.add_subplot(111)
ax1.set_title("Prediction Time")
ax1.set_xlabel("Number of points in Pool sets")
ax1.set_ylabel("Time")


for num_dims in dimensions_ar[:3]:
    #ax1.plot(num_pool_points_ar, time_simple_results[num_dims], ls = '--', label = "Simple, dim = {}".format(num_dims))
    ax1.plot(num_pool_points_ar, time_fast_results[num_dims], ls = '-.', label = "Fast, dim = {}".format(num_dims))
    ax1.plot(num_pool_points_ar, new_approach_time_fast_results[num_dims], 
             ls = '-.', label = "NEW Fast, dim = {}".format(num_dims))

plt.legend()
plt.show()
```
